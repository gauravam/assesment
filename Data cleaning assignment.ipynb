```
#data cleaning assignment
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',

'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```
df
```

#1

```
i=df[df["FlightNumber"].isnull()].index
```

```
df["FlightNumber"].fillna(method="ffill",inplace=True)
df["FlightNumber"][i]=df["FlightNumber"][i]+10
df["FlightNumber"]=df["FlightNumber"].astype(int)
```

```
df
```

#2

```
temp_df=df["From_To"].str.split('_',expand=True)
temp_df.rename({0:"From",1:"To"},axis=1,inplace=True)
```

```
temp_df
```

#3

```
temp_df["From"]=temp_df["From"].apply(str.capitalize)
```

```
temp_df["To"]=temp_df["To"].apply(str.capitalize)
```

```
temp_df
```

#4

```
del df["From_To"]
```

```
df_n=pd.concat([temp_df,df],axis=1)
df_n
```
